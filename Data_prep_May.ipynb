```python
def cleaned_merged(series):
    return pd.to_numeric(series
        .astype('string')
        .str.strip()
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.replace('%', '', regex=False),
        errors='coerce'
    ).fillna(0)
df_merged['MSF DC $'] = cleaned_merged(df_merged['MSF DC $'])
df_merged['MSF CC %'] = cleaned_merged(df_merged['MSF CC %'])
df_merged['Number of Transactions'] = cleaned_merged(df_merged['Number of Transactions'])
df_merged['Value of Transactions'] = cleaned_merged(df_merged['Value of Transactions'])
df_merged['Average of Transactions'] = cleaned_merged(df_merged['Average of Transactions'])
df_merged['COGS'] = cleaned_merged(df_merged['COGS'])
df_merged['MSF Margin'] = cleaned_merged(df_merged['MSF Margin'])
```

```python
df_merged = df_merged[df_merged['MSF CC %'] > 0 ]
df_merged.info()
```

```python
bins = [0, 0.9, 1.0, 1.2, 1.4, 1.65, float('inf')]
labels = ['0.0-0.9', '0.9-1.0', '1.0-1.2', '1.2-1.4', '1.4-1.65', '1.65+']
df_merged['Rate segment'] = pd.cut(
    df_merged['MSF CC %'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
df_merged.head()
```

```python
bins = [0, 50000, 100000, float('inf')]
labels = ['0-50K', '50K-100K', '100K+']
df_merged ['Tiers'] = pd.cut(
    df_merged['Value of Transactions'],
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=False
)
df_merged = df_merged[df_merged['Value of Transactions'] > 0]
```

```python
df_merged = df_merged.rename(
    columns={
        'Shipping State/Province' : 'State',
        'Shipping Zip' : 'PostCode'
    }
)
df_merged = df_merged.drop(
    columns=[
        'Trading Name', 'Mcc',
    ]
)
```

```python
df_merged['% Eftpos (non-Tap)'] = cleaned_merged(df_merged['% Eftpos (non-Tap)'])
df_merged['% Eftpos Tap'] = cleaned_merged(df_merged['% Eftpos Tap'])
df_merged['% Eftpos Total'] = cleaned_merged(df_merged['% Eftpos Total'])
df_merged['% Debit'] = cleaned_merged(df_merged['% Debit'])
df_merged['% Credit'] = cleaned_merged(df_merged['% Credit'])
df_merged['% International'] = cleaned_merged(df_merged['% International'])
df_merged['% Total Tap'] = cleaned_merged(df_merged['% Total Tap'])
```

```python
# Industry Analysis
df_industry = (
    df_merged.groupby('Business Industry')
    .agg(
        total_merchants=('Merchant ID', 'count'),
        total_ttv=('Value of Transactions', 'sum'),
        msf = ('COGS', 'sum'),
        total_MSF_margin=('MSF Margin', 'sum'),
        Average_of_Transactions=('Average of Transactions', 'mean'),
        Average_Eftpos_NonTap_transactions=('% Eftpos (non-Tap)', 'mean'),
        Average_Eftpos_tap_transactions=('% Eftpos Tap', 'mean'),
        Average_Eftpos_transactions=('% Eftpos Total', 'mean'),
        Average_Debit_transactions=('% Debit', 'mean'),
        Average_Credit_transactions=('% Credit', 'mean'),
        Average_International_transactions=('% International', 'mean'),
            Average_Total_tap_transactions=('% Total Tap', 'mean'),
        smartcharge_merchants=(
            'Price Model',
            lambda x: (x == 'SmartCharge').sum()
        ),
        SFR_merchants=(
            'Price Model',
            lambda x: (x == 'Simple rate').sum()
        )
    )
    .reset_index()
)
```

```python
 # Value must be sorted based on ttv otherwise cumulative ttv will be wrong
df_industry = df_industry.sort_values(
    by='total_ttv',
    ascending=False
)
# calculate cumulative TTV %
df_industry['ttv_cum_pct'] = (
    df_industry['total_ttv'].cumsum()
    / df_industry['total_ttv'].sum()
)

# calculate TTV %
df_industry['ttv_pct'] = (
    df_industry['total_ttv']
    / df_industry['total_ttv'].sum()
)

# cumulative MSF margin %
df_industry['MSF_margin_cum_pct'] = (
    df_industry['total_MSF_margin'].cumsum()
    / df_industry['total_MSF_margin'].sum()
)

# MSF margin %
df_industry['MSF_margin_pct'] = (
    df_industry['total_MSF_margin']
    / df_industry['total_MSF_margin'].sum()
)

# cumulative Smartcharge %
df_industry['smartcharge_cum_pct'] = (
   df_industry['smartcharge_merchants'].cumsum()
   / df_industry['smartcharge_merchants'].sum()
)

# Smartcharge %
df_industry['smartcharge_pct'] = (
   df_industry['smartcharge_merchants']
   / df_industry['smartcharge_merchants'].sum()
)


# cumulative SFR %
df_industry['SFR_cum_pct'] = (
  df_industry['SFR_merchants'].cumsum()
      / df_industry['SFR_merchants'].sum()
)

# SFR %
df_industry['SFR_pct'] = (
  df_industry['SFR_merchants']
      / df_industry['SFR_merchants'].sum()
)
```

```python
# 1. Total TTV per industry
industry_ttv = (
    df_merged.groupby('Business Industry')['Value of Transactions']
    .sum()
    .sort_values(ascending=False)
)

# 2. Top 15 industries
top15_ttv = industry_ttv.head(15)

# 3. Total TTV overall
total_ttv = industry_ttv.sum()

# 4. Percentage contribution of top 15
top15_percentage = ((top15_ttv.sum() / total_ttv)*100).round(0)

print(top15_percentage)
```

```python
result = pd.DataFrame({
    'Top 15 Industries TTV %': [(top15_ttv.sum() / total_ttv)],
    'Top 15 Industries TTV': [top15_ttv.sum()],
    'Total TTV': [total_ttv]
})
result.head()
```
